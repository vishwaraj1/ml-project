```python
x=df.iloc[:,:-1]
x.head(2)
```

```python
y=df['Class']
y.head(2)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=25)
x=pca.fit_transform(x)
```

```python
x.shape
```

```python
from sklearn.metrics import f1_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
def maxf1(clf,x,y):
    maxf=0
    rs=0
    for r_state in range(40,95):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=r_state,stratify=y)
        clf.fit(x_train,y_train)
        pred=clf.predict(x_test)
        tem=f1_score(y_test,pred)
        if tem>maxf:
            maxf=tem
            rs=r_state  
    print("maximum f1 score ",rs,'it is',maxf) 
    
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
maxf1(lr,x,y)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
svc=SVC()
parameters={'kernel':['linear','rbf','poly'],"C":[0.001,0.1,.5,.9,1,5,9.5,10,11]}
gd=GridSearchCV(svc,parameters,cv=5,scoring='f1')
gd.fit(x,y)
gd.best_params_
```

```python
svc=SVC(kernel='linear',C=5)
maxf1(svc,x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
parameters={"n_estimators":[20,30,40,50,60,70]}
gd=GridSearchCV(rf,parameters,cv=5,scoring='f1')
gd.fit(x,y)
gd.best_params_
```

```python
rf=RandomForestClassifier(n_estimators=30)
maxf1(rf,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = 86,test_size=0.20,stratify=y)
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
